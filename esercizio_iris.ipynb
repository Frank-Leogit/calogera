## Carico il datasets Iris

La variabile iris nel codice di esempio che ho fornito precedentemente è un oggetto restituito dalla funzione load_iris() della libreria scikit-learn. Non è un dizionario, ma un oggetto di tipo Bunch che contiene dati e metadati relativi al dataset Iris.

L'oggetto iris contiene i seguenti attributi:

- data: Un array NumPy che contiene le caratteristiche del dataset Iris (ovvero le misurazioni dei sepali e dei petali).
- target: Un array NumPy che contiene le etichette di classe del dataset Iris (ovvero le specie di fiori).
- target_names: Un array di stringhe che contiene i nomi delle classi nel dataset Iris.
- feature_names: Un array di stringhe che contiene i nomi delle caratteristiche nel dataset Iris.
DESCR: Una stringa che fornisce una descrizione del dataset Iris.

Inoltre, l'oggetto iris fornisce anche alcuni metodi per accedere ai dati, come ad esempio iris.keys() che restituisce un elenco delle chiavi degli attributi dell'oggetto iris, e iris.DESCR che restituisce la descrizione del dataset Iris.

Quindi, iris non è un dizionario ma un oggetto di tipo Bunch che contiene dati e metadati del dataset Iris fornito dalla libreria scikit-learn.

```python
iris = load_iris()
#Creazione del dataframe
df = pd.DataFrame(data=iris.data,columns=iris.feature_names)
#Aggiunta delle etichette di classe
df['target']= iris.target
#Cambio i nomi delle classi da valori a stringhe
df['target']=df['target'].replace({0:"Setosa",1:"Versicolor",2:"Virginica"})
df
```

## La normalizzazione delle statistiche 
è un passo comune nella preparazione dei dati per l'analisi e la modellazione dei dati, compresa l'applicazione di algoritmi di machine learning come la PCA (Principal Component Analysis). La normalizzazione è importante perché permette di mettere a confronto e trattare in modo equo le diverse caratteristiche (o feature) del dataset che possono avere scale di valori diverse.

Ciò può essere necessario perché alcune caratteristiche potrebbero avere un intervallo di valori molto più ampio rispetto ad altre. Ad esempio, nel caso del dataset Iris, le misurazioni dei sepali potrebbero variare da 1 a 10 cm, mentre le misurazioni dei petali potrebbero variare solo da 0,1 a 3 cm. Se non si normalizzano le caratteristiche, quelle con un intervallo di valori più ampio avranno un impatto maggiore sulle analisi o sui modelli rispetto a quelle con un intervallo di valori più stretto, semplicemente a causa della loro scala di valori più ampia.

La normalizzazione delle caratteristiche riduce questa discrepanza nella scala dei valori, consentendo di trattare tutte le caratteristiche in modo uniforme e di evitare che alcune caratteristiche dominino altre nel processo di analisi o modellazione. Inoltre, alcuni algoritmi di machine learning, come la PCA, sono sensibili alla scala delle caratteristiche e possono essere influenzati negativamente da caratteristiche con scale di valori molto diverse.

La normalizzazione può essere effettuata utilizzando diverse tecniche, come ad esempio la Standardizzazione (o Z-score normalization) o la Normalizzazione Min-Max, che portano i valori delle caratteristiche a una scala comune o ad un intervallo specifico di valori, rispettivamente. La scelta della tecnica di normalizzazione dipende dal contesto dell'analisi e dall'algoritmo di machine learning utilizzato. Nella maggior parte dei casi, la normalizzazione delle statistiche è un passo importante per garantire risultati accurati e consistenti nelle analisi e nei modelli basati su dati.

```python
from sklearn.preprocessing import StandardScaler
df_standardized = df.copy()
df_standardized[data]= StandardScaler().fit_transform(df[data])
```

```python
df_standardized[data].describe().T
```

```python
df_standardized[data].shape
```

```python
from sklearn.decomposition import PCA

myPCA = PCA().fit(df_standardized[data])

print(myPCA.explained_variance_ratio_)
print(myPCA.explained_variance_ratio_.cumsum())
```
